```python
import pandas as pd 
import numpy as np
import sklearn.tree as tree
import sklearn.model_selection as ms
import sklearn.ensemble as ens
import sklearn.metrics as met
```

```python
rf = ens.RandomForestClassifier(n_estimators=100) #100 buah Decision Tree
df = pd.read_csv("adult_data.csv")
df.head(10)
```

```python
df.dropna(inplace=True) #menghilangkan missing value
```

```python
#mengubah feature nominal menjadi numerik
import sklearn.preprocessing as prep
workclass = prep.LabelEncoder()
workclass.fit(df['workclass'])
df['workclass_code'] = workclass.transform(df['workclass'])
df[['workclass', 'workclass_code']].head(5)
```

```python
gender = prep.LabelEncoder()
gender.fit(df['gender'])
df['gender_code'] = gender.transform(df['gender'])
df[['gender', 'gender_code']].head(5)
```

```python
education = prep.LabelEncoder()
education.fit(df['education'])
df['education_code'] = education.transform(df['education'])
df[['education', 'education_code']].head(5)
```

```python
marital = prep.LabelEncoder()
marital.fit(df['marital-status'])
df['marital_code'] = marital.transform(df['marital-status'])
df[['marital-status', 'marital_code']].head(5)
```

```python
occupation = prep.LabelEncoder()
occupation.fit(df['occupation'])
df['occupation_code'] = occupation.transform(df['occupation'])
df[['occupation', 'occupation_code']].head(5)
```

```python
relationship = prep.LabelEncoder()
relationship.fit(df['relationship'])
df['relationship_code'] = relationship.transform(df['relationship'])
df[['relationship', 'relationship_code']].head(5)
```

```python
race = prep.LabelEncoder()
race.fit(df['race'])
df['race_code'] = race.transform(df['race'])
df[['race', 'race_code']].head(5)
```

```python
country = prep.LabelEncoder()
country.fit(df['native-country'])
df['country_code'] = country.transform(df['native-country'])
df[['native-country', 'country_code']].head(5)
```

```python
df.head()
```

```python
X = df.drop(['x', 'fnlwgt', 'workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'native-country', 'income'], axis=1)
y = df['income']
```

```python
X_train, X_test, y_train, y_test = ms.train_test_split(X, y, test_size = 0.2) #membagi data training dan data testing
```

```python
rf.fit(X_train, y_train) #modelling
y_prediksi = rf.predict(X_test)
print('akurasi = ', met.accuracy_score(y_test, y_prediksi))
print(met.classification_report(y_test, y_prediksi))
print(rf.feature_importances_)
```
